EXPLORATORY DATA ANALYSIS

```python
# Histograms for all numerical features
import matplotlib.pyplot as plt
data.hist(figsize=(10, 10))
plt.show()
```

```python
# Box plots for checking outliers in numerical features
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure()
    sns.boxplot(x=data[column])
    plt.title(column)
    plt.show()
```

```python
# Exclude non-numeric columns from the correlation matrix calculation
numeric_data = data.select_dtypes(include=[np.number])  # Only select columns with numeric data types

# Now you can safely calculate the correlation matrix
corr_matrix = numeric_data.corr()
```

```python
# Plotting the correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.show()
```
